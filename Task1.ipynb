```python
### Task 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"E:\cleaned_stroke_data.csv")
print("Dataset loaded successfully!\n")
print(df.head())
```

```python
print(df.describe())
```

```python
numeric_cols = ['age', 'avg_glucose_level', 'bmi']
for col in numeric_cols:
    print(f"\nColumn: {col}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Min: {df[col].min():.2f}")
    print(f"Max: {df[col].max():.2f}")
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    print(f"\n Outliers detected in {col}: {len(outliers)}")
```

```python
print("\n Correlation Matrix:")
print(df.corr(numeric_only=True))
```

```python
# Histogram (distribution of BMI)
plt.figure(figsize=(6,4))
sns.histplot(df['bmi'], bins=30, color='skyblue')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar chart - Average glucose level by gender
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='avg_glucose_level', data=df)
plt.title("Average Glucose Level by Gender")
plt.xlabel("Gender")
plt.ylabel("Avg Glucose Level")
plt.show()
```

```python
#Line graph - Age vs Average Glucose Level
plt.figure(figsize=(8,5))
sns.lineplot(x='age', y='avg_glucose_level', data=df, color='orange')
plt.title("Age vs Average Glucose Level")
plt.xlabel("Age")
plt.ylabel("Avg Glucose Level")
plt.show()
```

```python
#Boxplot - BMI Outliers by Stroke Status
plt.figure(figsize=(6,4))
sns.boxplot(x='stroke', y='bmi', data=df)
plt.title("BMI Distribution by Stroke Status")
plt.xlabel("Stroke (0 = No, 1 = Yes)")
plt.ylabel("BMI")
plt.show()
```
